```python
#4
N=11
h=1/(N-1)
x=zero_vector(RR,N)
y=zero_vector(RR,N)
y[0]=1
x[0]=0
ye=zero_vector(RR,N)
ye[0]=1
for i in range(1,N):
    x[i]=x[i-1]+h
    y[i]=y[i-1]+h*2*x[i-1]*y[i-1]
    ye[i]=exp(x[i]^2)
    
ly=line(Line(range(N),y,{}),marker='o')
lye=line(Line(range(N),ye,{}),color='cyan',marker='o')
show(ly+lye)
```
