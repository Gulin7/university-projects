```python
#10
t=var('t')
x=function('x')(t)
sol1 = desolve(diff(x,t,2)+x==t, x, ics=[0,5,0])
sol2 = desolve(diff(x,t,2)+x==pi-t, x, ics=[pi/2,sol1(pi/2),diff(sol1,t).subs(t=pi/2)])
sol3 = desolve(diff(x,t,2)+x==0, x, ics=[pi,sol2(pi),diff(sol2,t).subs(t=pi)])
f = piecewise([((0,pi/2),sol1),((pi/2,pi),sol2),((pi,infinity),sol3)])
plot(f,0,10)
```
