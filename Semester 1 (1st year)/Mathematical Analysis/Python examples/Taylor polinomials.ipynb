# Taylor polynomials for the exponential

Consider the exponential function $e^x$ and its Taylor series $e^x = \displaystyle\sum_{k=0}^{\infty} \frac{x^k}{k!} = 1+x+\frac{x^2}{2} + \frac{x^3}{6}+\ldots$.

We will plot below the first Taylor polynomials $T_n = \displaystyle\sum_{k=0}^{n} \frac{x^k}{k!}$.

```python
import matplotlib.pyplot as plt
import numpy as np

# settings for plot
plt.rcParams['text.usetex'] = True
plt.rcParams["figure.figsize"] = (10,10)

# define the exponential
x = np.linspace(-2,2,100)
exp_x = np.exp(x)

# define Taylor's polynomials
n = 6
T = np.zeros([n, len(x)])
T[0:] = np.ones(len(x))
for i in range(1,n):
    T[i,:] = T[i-1,:] + x**i/np.math.factorial(i)

# plot the exponential
plt.plot(x, exp_x, "k", label=r"$e^x$")

# plot Taylor's polynomial
linestyle = ["--",""]
for i in range(1,5):
    plt.plot(x, T[i], "--", label=r"$T_"+str(i)+"(x)$", )

# make the plot nicer
plt.xlabel("x")
plt.ylabel("y")
plt.legend(fontsize=16)
plt.xlim((-2,2))
plt.xticks(range(-4, 5, 1))
plt.yticks(range(-1, 8, 1))
plt.grid()
plt.axis("equal")
plt.show()
```
